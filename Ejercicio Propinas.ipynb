### El mesero trabajó Jueves, Viernes, Sábado y Domingo

```python
df_propinas.groupby("day")[["tip"]].sum()
```

### Como no tenemos las fechas de los días no puedo saber exactamente cuántas propinas recaudó en una sola fecha.

### Pregunta, quién deja más propinas? Fumadores o no fumadores?

```python
df_propinas.groupby("smoker")[["tip"]].sum()
```

### Los fumadores dejaron más propinas que los no fumadores.
### Pero esto puede depender de la cantidad de fumadores.

```python
df_propinas["smoker"].value_counts()
```

### En proporción son casi el doble de no fumadores.
### Pregunta, en promedio cuánta propina deja cada uno?

```python
df_propinas.groupby("smoker")[["tip"]].mean()
```

### En proporción dejan casi lo mismo, aunque los fumadores dejan poquito más
