```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_pelis = pd.read_excel("movies-2.xlsx", sheet_name="movies", header=0, names=["pelicula", "genero", "estudio", "score", "profitability", "criticas", "recaudación", "año"])
df_pelis
```

```python
 len(df_pelis)
```

### Hat 77 películas en el dataframe

```python
df_pelis.groupby("genero")[["pelicula"]].count()
```

```python
df_pelis["genero"] = df_pelis["genero"].str.lower().str.strip()
```

```python
# Ahora hagamos las correciones
correciones = {
    "comdy":"comedy",
    "romence":"romance"
}
# Aplicamos el diccionario

df_pelis["genero"] = df_pelis["genero"].replace(correciones)
```

```python
df_pelis.groupby("genero")[["pelicula"]].count()
```

```python
plt.hist(df_pelis["genero"], bins=6, color="blue",edgecolor="black")
plt.title("Géneros de películas")
plt.xlabel("Género")
plt.ylabel("Freq")
plt.show()
```
