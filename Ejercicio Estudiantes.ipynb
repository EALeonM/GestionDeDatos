```python
variables = ['stress', 'critical', 'grade', 'regularity']
corr = df_estudiantes[variables].corr()
corr
```

#### A estrés y habilidades de pensamiento crítico: Correlación casi nula
#### B calificación y los patrones de regularidad del alumno a lo largo del curso: Correlación Media positiva
#### C calificación y pensamiento crítico: Correlación Media positiva, pero menor a la regularidad
#### D calificación y estrés: Correlación Media negativa

```python
fig, axs = plt.subplots(4,1, figsize=(15, 15))
plt.suptitle("Coeficiente de correlación")

axs[0].scatter(corr["stress"], corr["critical"], color = "navy")
axs[0].set_title("Correlación Estrés(x) Critical(y)")

axs[1].scatter(corr["grade"], corr["regularity"], color = "red")
axs[1].set_title("Correlación B grade(x) regularity(y)")

axs[2].scatter(corr["grade"], corr["critical"], color = "orange")
axs[2].set_title("Correlación C grade(x) critical(y)")

axs[3].scatter(corr["grade"], corr["stress"], color = "purple")
axs[3].set_title("Correlación D grade(x) stress(y)")

plt.tight_layout
plt.show()
```

#### 1 Hay una alta correlación entre la regularidad y buenas calificaciones sé regular
#### 2 Ten pensamiento crítico, también hay correlación
#### 3 No te estreses de más, el estrés baja el desempeño
