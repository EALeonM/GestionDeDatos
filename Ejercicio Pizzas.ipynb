### El precio más caro con descuento para pizza es de 264 pesos, para bebida de 35 y de botana 104

```python
max_pizza= df_pizza[df_pizza['CLASIFICACIÓN'] == 'pizza']['PRE_DESC'].max()
max_bebida= df_pizza[df_pizza['CLASIFICACIÓN'] == 'bebidas']['PRE_DESC'].max()
max_botana= df_pizza[df_pizza['CLASIFICACIÓN'] == 'botana']['PRE_DESC'].max()
```

```python
df_resultado = df_pizza.loc[
    ((df_pizza['CLASIFICACIÓN'] == 'pizza')   & (df_pizza['PRE_DESC'] == max_pizza)) |
    ((df_pizza['CLASIFICACIÓN'] == 'bebidas') & (df_pizza['PRE_DESC'] == max_bebida)) |
    ((df_pizza['CLASIFICACIÓN'] == 'botana')  & (df_pizza['PRE_DESC'] == max_botana)), ["NOMBRE", "PRE_DESC"]
]

df_resultado.drop_duplicates(subset='NOMBRE')
```

```python
df_pizza.groupby("CLASIFICACIÓN").agg({"PRECIO_U":["sum","mean"], "CANT": ["sum","mean"], "TOTAL": ["sum","mean"]})
```

```python
df_pizza.groupby("CLASIFICACIÓN")[["CANT"]].count()
```

### Esto si se refiere a de los 1200 pedidos de los que se tienen registro, cuantos fueron de bebidas, cuantos de botana y cuantos de pizza.

```python
plt.hist(df_pizza["CANT"],bins=5,color='coral',edgecolor='black')
plt.title("Frecuencia de cantidad de pedidos")
plt.xlabel("Cantidad de Pedidos")
plt.ylabel("Frecuencia")
plt.show()
```

```python
df_alitas = df_pizza.loc[
    ((df_pizza["CLASIFICACIÓN"]=="botana") & (df_pizza["NOMBRE"]=="alitas")), ["NOMBRE", "CANT"]
]
df_alitas
```

### Se hicieron 106 pedidos de alitas

```python
df_ordenada = df_pizza.sort_values(by=["PRECIO_U"], ascending=True)
df_ordenada.head()
```

```python
df_promedios = df_pizza.groupby("CLASIFICACIÓN")[["TOTAL"]].mean()
df_promedios
```

```python
plt.bar(df_promedios.index, df_promedios["TOTAL"], color="coral")
plt.title("Promedio de ingreso por producto ordenado")
plt.xlabel("Producto")
plt.ylabel("Promedio")
plt.show()
```
