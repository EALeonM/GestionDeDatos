#### Ninguna de nuestras variables tienen una correlación alta como para excluirlas.

```python
X = df_vinos[["WinterRain", "AGST", "HarvestRain", "Age"]]
y = df_vinos["Price"]
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
a0 = lr.intercept_
a1 = lr.coef_[0]
a2 = lr.coef_[1]
a3 = lr.coef_[2]
a4 = lr.coef_[3]
print("La ecuación es: y =", a0,"+ ", a1,"x1 + ", a2,"x2 + ", a3,"x3 + ", a4,"x4")
```

```python
R2 = lr.score(X, y)
R2
```

#### La regresión es bastante buena.

```python
dic = {'WinterRain': [600], 'AGST': [15], 'HarvestRain': [150], 'Age': [15]}
df_predic = pd.DataFrame(dic)
df_predic
```

```python
price_predic = lr.predict(df_predic)
price_predic
```

```python
df_predic['Price'] = price_predic
df_predic
```

```python
scaler = StandardScaler()
```

```python
# Separando la normalización de X en fit y transform (de métodos en vez de 1)
X_fit = scaler.fit(X)
X_trans = pd.DataFrame(X_fit.transform(X), columns=X.columns)
X_trans
```

```python
#También necesitamos normalizar el df de prueba
dic = {'WinterRain': [600], 'AGST': [15], 'HarvestRain': [150], 'Age': [15]}
df_predic1 = pd.DataFrame(dic)
df_predic1
```

```python
df_predic_norm = pd.DataFrame(X_fit.transform(df_predic1), columns=df_predic1.columns)
df_predic_norm
```

```python
# Separando la normalización de y en fit y transform (2 métodos en vez de 1)
y_fit = scaler.fit(df_vinos[['Price']])
y_trans = pd.DataFrame(X_fit.transform(df_vinos[["Price"]]), columns=df_vinos[["Price"]].columns)
# Usamos en la transformación de y X_fit para que la columna price esté normalizada con x
```

```python
# Hacemos una nueva regresión con los valores normalizados
reg = LinearRegression()
reg.fit(X_trans, y_trans)
```

```python
# Calculamos los coeficientes, el intercepto y R2
coef_n = reg.coef_
intercept_n = reg.intercept_
r2_n = reg.score(X_trans, y_trans)
print("Coeficientes: ", coef_n, "Intercept: ", intercept_n, "R2: ", r2_n)
```

```python
#Ahora podemos hacer la predicción
res_norm = reg.predict(df_predic_norm)
res_norm
```

```python
# El precio real deberá ser transformado usando el objeto scaler
precio = X_fit.inverse_transform(res_norm)
precio
```
