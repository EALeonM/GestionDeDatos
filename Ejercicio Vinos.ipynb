```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
df_vinos = pd.read_csv("wines-1-1.csv")
df_vinos.head()
```

#### Podemos predecir la calidad del vino? según su precio?
#### Nuestra y será el precio.
#### Nuestras x serán WinterRain, AGST, HarvestRain, y Age
#### Son las más determinantes debido a que son las que más información nos dan sobre la calidad del vino.

```python
df_vinos[["WinterRain", "AGST", "HarvestRain", "Age"]].corr()
```

#### Ninguna de nuestras variables tienen una correlación alta como para excluirlas.

```python
X = df_vinos[["WinterRain", "AGST", "HarvestRain", "Age"]]
y = df_vinos["Price"]
```

```python
lr = LinearRegression()
lr.fit(X, y)
```

```python
a0 = lr.intercept_
a1 = lr.coef_[0]
a2 = lr.coef_[1]
a3 = lr.coef_[2]
a4 = lr.coef_[3]
print("La ecuación es: y =", a0,"+ ", a1,"x1 + ", a2,"x2 + ", a3,"x3 + ", a4,"x4")
```

```python
R2 = lr.score(X, y)
R2
```

#### La regresión es bastante buena.

```python
dic = {'WinterRain': [600], 'AGST': [15], 'HarvestRain': [150], 'Age': [15]}
df_predic = pd.DataFrame(dic)
df_predic
```

```python
price_predic = lr.predict(df_predic)
price_predic
```

```python
df_predic['Price'] = price_predic
df_predic
```
