```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_ine = pd.read_csv("INE_ENTIDAD_2020-3.CSV", index_col = "NOM_ENT ", encoding = 'latin-1')
df_ine.head()
```

```python
df_ine.info()
```

```python
df_ine.columns
```

```python
df_mod_ine = pd.DataFrame(df_ine.loc[:,["POBTOT", "POBFEM", "POBMAS"]])
df_mod_ine.head()
```

```python
df_mod_ine.dtypes
```

```python
df_mod_ine.sum()
```

POBTOT es el total de la población registrada
POBFEM es el total de la población femenina registrada
POBMAS es el total de la población masculina registrada

```python
df_pobtot = pd.DataFrame(df_mod_ine.loc[:,"POBTOT"])
df_pobtot
```

```python
df_pobtot.loc[df_pobtot.idxmin()]
# El renglón con menos personas registradas es: "731391"
```

```python
df_pobtot.idxmin()
# Colima es el Estado con el menor registro de personas
```

```python
df_pobtot.loc[df_pobtot.idxmax()]
# El renglón con más personas registradas es: "16992418"
```

```python
df_pobtot.idxmax()
# El estado con más personas registradas es: "México"
```

```python
df_pobtot.describe()
# Sí, coincide el máximo y el mínimo
```

```python
df_pobfem = pd.DataFrame(df_mod_ine.loc[:,"POBFEM"])
df_pobfem
```

```python
plt.barh(df_pobfem.index , df_pobfem["POBFEM"], color = 'coral')
plt.title("Población de mujeres por Estado")
plt.xlabel("Población")
plt.ylabel("Estado")
plt.show()
```

```python
df_pobmas = pd.DataFrame(df_mod_ine.loc[:,"POBMAS"])
df_pobmas
```

```python
plt.hist(df_pobmas["POBMAS"], bins = 10, color = 'coral', edgecolor = 'black')
plt.title("Distribución de la población masculina por estado")
plt.xlabel("Población masculina")
plt.ylabel("Frecuencia")
plt.show()
```
