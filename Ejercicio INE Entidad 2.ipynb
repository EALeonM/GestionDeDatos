```python
# Vamos a crear la nueva serie con relaciÃ³n mujeres/hombres
rel_fem_mas = pd.Series(df_ine_mod["POBFEM"]/df_ine_mod["POBMAS"], name = "RelaciÃ³n FEM/MAS")
rel_fem_mas.head()
```

```python
rel_fem_mas_mod = pd.Series(rel_fem_mas-1)
rel_fem_mas_mod.head()
```

```python
plt.barh(rel_fem_mas_mod.index, rel_fem_mas_mod, color = "coral")
plt.title("RelaciÃ³n mujeres/hombres por entidad")
plt.xlabel("RelaciÃ³n")
plt.ylabel("Entidad")
plt.show()
```

```python
rel_max = rel_fem_mas_mod.idxmax()
print(rel_max)
```

```python
valor_rel_max = rel_fem_mas_mod.max()
print(valor_rel_max)
```
